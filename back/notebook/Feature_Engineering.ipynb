# Feature Engineering

# Load Clean Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/cleaned_dataset.csv")

print(df.shape)
print(df.columns)
```

# Separate Features & Target

```python
X = df.drop('target', axis=1)
y = df['target']

print("X shape:", X.shape)
print("y shape:", y.shape)
```

# Identify Feature Types

```python
categorical_cols = [
    'sex',
    'cp',
    'fbs',
    'restecg',
    'exang',
    'slope',
    'ca',
    'thal'
]

numerical_cols = [
    'age',
    'trestbps',
    'chol',
    'thalach',
    'oldpeak'
]
```

# Create New Engineered Features

```python
# Blood pressure to cholesterol ratio
X['bp_chol_ratio'] = X['trestbps'] / (X['chol'] + 1)

# Age grouped
X['age_group'] = pd.cut(
    X['age'],
    bins=[29, 40, 55, 70, 100],
    labels=[0,1,2,3]
)

# High cholesterol flag
X['high_chol'] = (X['chol'] > 240).astype(int)
```

# Update Feature Lists After Engineering

```python
categorical_cols.append('age_group')
categorical_cols.append('high_chol')

numerical_cols.append('bp_chol_ratio')
```

# Apply Scaling + Encoding Properly

Numerical Features

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(drop='first'), categorical_cols)
    ]
)

model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

model.fit(X, y)

print("Feature Engineering + Training Successful ✅")
```
